## Visualizations

```python
#importing python libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from recommendations import df_books, df_ratings, df_users
```

### Number of books published yearly

```python
# count number of books published each year
books_per_year = df_books["Year-Of-Publication"].value_counts().sort_index()

plt.bar(books_per_year.index, books_per_year.values)
plt.xlim(1950, 2021)
plt.xlabel("Year")
plt.ylabel("Number of Books Published")
plt.title("Number of Books Published Each Year")
plt.show()
```

### Number of books published by top 15 authors

```python
# Count the number of books by author
author_counts = df_books["Book-Author"].value_counts()[:15]

# Create a bar plot
plt.bar(author_counts.index, author_counts.values)
plt.xticks(rotation=90)
plt.xlabel("Author")
plt.ylabel("Number of Books")
plt.title("Number of Books by Top 15 Authors")
plt.show()
```

### Number of books published by top 15 publishers

```python
# Count the number of books by publisher
publisher_counts = df_books["Publisher"].value_counts()[:15]

# Create a bar plot
plt.bar(publisher_counts.index, author_counts.values)
plt.xticks(rotation=90)
plt.xlabel("Publisher")
plt.ylabel("Number of Books")
plt.title("Number of Books by Top 15 Publishers")
plt.show()
```

### Number of readers from top 15 cities

```python
city_counts = df_users["City"].value_counts()

plt.bar(city_counts.index[0:15], city_counts.values[0:15])
plt.xticks(rotation=90)
plt.xlabel("City")
plt.ylabel("Number of Readers")
plt.title("Number of Readers from Each City (Top 15)")
plt.show()
```

### Number of readers from top 15 states

```python
state_counts = df_users["State"].value_counts()

plt.bar(state_counts.index[0:15], state_counts.values[0:15])
plt.xticks(rotation=90)
plt.xlabel("State")
plt.ylabel("Number of Readers")
plt.title("Number of Readers from Each State (Top 15)")
plt.show()
```

```python
country_counts = df_users["Country"].value_counts()

plt.bar(country_counts.index[0:15], country_counts.values[0:15])
plt.xticks(rotation=90)
plt.xlabel("Country")
plt.ylabel("Number of Readers")
plt.title("Number of Readers from Each Country (Top 15)")
plt.show()
```
